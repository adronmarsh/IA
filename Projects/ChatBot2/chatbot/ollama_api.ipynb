```python
import ollama
response = ollama.chat(model='mistral', messages=[
  {
    'role': 'user',
    'content': 'Why is the sky blue?',
  },
])
print(response['message']['content'])
```

```python
import ollama

stream = ollama.chat(
    model='mistral',
    messages=[{'role': 'user', 'content': 'Why is the sky blue?'}],
    stream=True,
)

for chunk in stream:
  print(chunk['message']['content'], end='', flush=True)
```

```python
modelfile='''
FROM mistral
SYSTEM You are mario from super mario bros.
'''

ollama.create(model='mario', modelfile=modelfile)
```

```python
stream = ollama.chat(model='mario', messages=[{'role': 'user', 'content': 'Who are you?'}], stream=True)

for chunk in stream:
  print(chunk['message']['content'], end='', flush=True)
```

```python
respuesta_completa = ""
stream = ollama.chat(model='mario', messages=[{'role': 'user', 'content': 'Who are you?'}], stream=True)

for chunk in stream:
    respuesta_completa += chunk['message']['content']
```

```python
print(respuesta_completa)
```
