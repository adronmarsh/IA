En este caso se han eliminado.

## Fase 3 - Anonimizacin

Tcnicas:
- Supresin de registros --> Eliminar la fila entera
- Supresin de atributos --> Eliminar el atributo en concreto
- Enmascaramiento de caracteres --> 357703 --> 357XXX
- Generalizacin --> Avenida de Madrid,22 1B --> Avenida de Madrid
- Perturbacin de datos --> Modificar los valores utilizando tcnicas de redondeo o agregacin de ruido
- Seudonimizacin --> Sustituir los datos por valores inventados (parecido al hash) Ej: Carlos Muoz -> Alm16L 3Las4
- Intercambio de datos --> Intercambiar los datos entre los registros. Ej: Carlos -> Verde, Mario -> Azul --> Carlos -> Azul, Mario -> Verde


## Generalizacin y enmascaramiento

- Persona --> Enmascarar. Ej: 357703 --> 357XXX

- Edad --> Generalizar. Ej: 24 --> 20-30

- Direccin --> Generalizar. Ej: Avenida de Madrid,22 1B --> Avenida de Madrid
