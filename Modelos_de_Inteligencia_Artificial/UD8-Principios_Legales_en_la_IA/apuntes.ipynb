## Fase 2 - Desidentificar

Lo normal es elimninar los directos aunque también se puede asignar un seudónimo a cada registro si se desea 
conservar la capacidad de vincular el registro de datos desidentificados al 
registro original.

En este caso se han eliminado.

## Fase 3 - Anonimización

Técnicas:
- Supresión de registros --> Eliminar la fila entera
- Supresión de atributos --> Eliminar el atributo en concreto
- Enmascaramiento de caracteres --> 357703 --> 357XXX
- Generalización --> Avenida de Madrid,22 1ºB --> Avenida de Madrid
- Perturbación de datos --> Modificar los valores utilizando técnicas de redondeo o agregación de ruido
- Seudonimización --> Sustituir los datos por valores inventados (parecido al hash) Ej: Carlos Muñoz -> Alm16L 3Las4
- Intercambio de datos --> Intercambiar los datos entre los registros. Ej: Carlos -> Verde, Mario -> Azul --> Carlos -> Azul, Mario -> Verde


## Generalización y enmascaramiento

- Persona --> Enmascarar. Ej: 357703 --> 357XXX

- Edad --> Generalizar. Ej: 24 --> 20-30

- Dirección --> Generalizar. Ej: Avenida de Madrid,22 1ºB --> Avenida de Madrid
