### ¿Qué es Blockchain?
Blockcahin es una red distribuida



**Componentes**
- Nodos
- Bloques
- Transacciones
- Smart Contracts

### Nodos
- Tiene una copia completa de los datos de la blockchain

- Funciones:
  - Validación
  - Propagación
  - Almacenamiento
  - Consenso

### Bloques
- Contienen información en forma de transacciones
- Están enlazados
- Ordenados temporalmente

### Tecnologías
- Criptografía
  - Firmas con clave pública/privada
  - Hashing
- Consenso
  - Proof of Work
  - Prook of Stake
- Smart Contracts
- Hash
