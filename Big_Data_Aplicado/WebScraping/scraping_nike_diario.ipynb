```python
import datetime
import os
import csv
import time
def CreateCVSFile(elements_list,urlfile):
    today=  time.strftime("%y%m%d")
    filename =  urlfile+str(today)+'.csv'
    with open(filename, "w",newline='', encoding='utf-8') as stream:
        writer = csv.writer(stream)
        writer.writerows(elements_list)
```

 # 1a. A partir del site map construimos un array de posibles paginas a visitar 

 la variable lista_paginas tiene todas la urls del site map

```python


#sitemap inicial en castellano  --> https://www.nike.com/sitemap-v2-landingpage-es-es.xml
sitemap_url ="https://www.nike.com/sitemap-v2-landingpage-es-es.xml"

#tag= "xhtml:link"
#attribute_lang ="hreflang"
#attribute_value="href"

tag="url"

lista_paginas_desde_site_map=[]


#funcion que recorre un sitemap que recibe c√≥mo par√°metro y devuelve la lista de urls
def listadoPaginasSiteMap(page_site_map,tag):
    page_site_map = requests.get(sitemap_url)
    sitemap = BeautifulSoup(page_site_map.content, "lxml")
    tags = sitemap.find_all(tag)
    aux_lista_paginas=[]
    for pagina_site in tags:
        #print (pagina_site.get_text().strip())
        aux_lista_paginas.extend([pagina_site.loc.get_text().strip()])
    return aux_lista_paginas

lista_paginas_desde_site_map=listadoPaginasSiteMap(sitemap_url,tag)

print (lista_paginas_desde_site_map)
```

# 1.b a partir del menu se obtienen las urls de las diferetnes categorias

```python
#iter

#<a role="menuitem" aria-label="main-menu, Hombre, Zapatillas" tabindex="-1" data-nav="1,1,0" data-type="click_navShoppingMenu" data-path="men:shoes" 
#href="https://www.nike.com/es/w/hombre-zapatillas-nik1zy7ok" class="pre-menu-item is-static is-header headline-5" data-pre="ILink">Zapatillas</a>


import requests

lista_paginas_desde_menu=[]

url_home = "https://www.nike.com/es/"
def listadoPaginasMenu(urlpage):
    page = requests.get(urlpage)
    page_soup = BeautifulSoup(page.content)

    aux_lista_paginas=[]
    for menu_item in page_soup.find_all("a", {"role": "menuitem"}):
        aux_lista_paginas.extend([menu_item.get("href")])
    return aux_lista_paginas

lista_paginas_desde_menu =listadoPaginasMenu(url_home)
print (lista_paginas_desde_menu)
```

```python
# para cada pagina de la lista hacemos la peticion y buscamos la etiqueta de producto y lo a√±adimos a la lista 




#recupera todos los productos de una determinada pagina
def recuperaProductosPagina(url):
    page = requests.get(url)
    page_soup = BeautifulSoup(page.content)

        ## probamos a recuperar los productos desde una categoria con listado


    listaproductos_pagina=[] # lista en la cual guardamos la informaci√≥n de todos los productos



    #recuperamos todos los productos de la lista   (en caso de ser una pagina de lista)

    #<div class="product-card__body
    for item_product in page_soup.find_all("div", {"class": "product-card__body"}):
        
        nombre_producto= item_product.a.get_text().strip()
        #<div class="product-card__subtitle" role="link">Zapatillas - Mujer</div>
        categoria=item_product.find("div", {"class": "product-card__subtitle"}).get_text().strip()
	    

        #<div class="product-price es__styling is--current-price css-11s12ax" data-testid="product-price" role="link">119,99&nbsp;‚Ç¨</div>
        #precio:float
	    #img_url_cabecera:str
	    #url_producto
		

      
        oProducto = Producto( nombre_producto,categoria,None,None,None,None,None,None,None,None)
      
        listaproductos_pagina.append(oProducto) 
    return listaproductos_pagina
    
  









num_paginas_max  = 30
num_pag_recorridas =0

listaproductos=[] # lista en la cual guardamos la informaci√≥n de todos los productos de todas las paginas


#iteramos por todas las listas de productos
for pagina_url in lista_paginas_desde_menu:
    print( "pagina_url: " +str(num_pag_recorridas)+ " " + pagina_url)
    productos_recuperados =recuperaProductosPagina(pagina_url) 
    listaproductos.extend(productos_recuperados)
    ##cada lista de productos puede tener a su ver paginacion
    #paginacion =2
    #while productos_recuperados != []:
    #    nueva_url = pagina_url + "?searchParams=&sort=&view=&page=" + str(paginacion)
    #    paginacion = paginacion + 1
    #    productos_recuperados =recuperaProductosPagina(nueva_url) 
    #    listaproductos.extend(productos_recuperados)
    #    print(paginacion)
    num_pag_recorridas = num_pag_recorridas +1
    if num_pag_recorridas >num_paginas_max:
        break


```
