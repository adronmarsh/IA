**0. Librerias necesarias para el laboratorio.**

```python
import pandas as pd
import numpy as np

 
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.feature_selection import SelectKBest
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

```

**1. Análisis descriptivo de los datos:**

```python
# Importar el csv para poder leer los datos
df = pd.read_csv('lab1-dataset.csv', sep=",")
df
```

**2. Tratamiento de missing:**

```python
# Contar los null que hay en cada columnna (por curiosidad)
df.isnull().sum()
```

```python
# Sacar la moda de cada columna
moda = df.mode().iloc[0]
moda
```

```python
# Rellenar los null con la moda
df = df.fillna(moda)
df
```

**3. Determinar el conjunto de modelización y el de validación.**

Buscamos dividir nuestros datos para que el 75 % de los datos sean para entrenamiento y el 25 % para test.


```python
# Dividir los datos en conjunto de entrenamiento y de test
from sklearn.model_selection import train_test_split

X = df.drop("classname", axis=1)  # Características
```

```python
y = df["classname"]  # Objetivo
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.289, random_state=37)
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=37)
print("Ejemplos usados para entrenar", len(X_train))
print("Ejemplos usados para test", len(X_test))
```

Nos ponemos a entrenar nuestros datos, para ello vamos a utilizar Multinomial Naive Bayes al ser datos discretos.

```python
from sklearn.naive_bayes import MultinomialNB
naive_bayes = MultinomialNB() # Crear el clasificador
```

```python
naive_bayes.fit(X_train, y_train) # Entrenar al modelo
```

Por último vamos a predecir utilizando los datos test.

**4. Calcular las métricas de evaluación de ajuste adecuadas.**

```python
# Realizar predicciones en el conjunto de prueba
y_pred = naive_bayes.predict(X_test)

# Calcular la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)
print("Precisión del modelo:", accuracy)
```
