1. Máximo común divisor de dos números:

```python
# Pedimos al usuario que ingrese los numeros
num1 = int(input("Ingrese el primer numero\n:"))
num2 = int(input("Ingrese el segundo numero\n:"))
# Seleccionamos el mayor y el menor y los asignamos
# a las variables "a" y "b" respectivamente
maximo = max(num1, num2)
minimo = min(num1, num2)
# Realizamos el ciclo encargado de hacer las iteraciones
while minimo!=0:
    resultado = minimo
    minimo = maximo%minimo
    maximo = resultado
# Mostramos el resultado en pantalla
print('El M.C.D. entre {0} y {1} es: {2}'.format(num1, num2, resultado))
```

2. Mínimo común múltiplo de dos números.

```python
 """ Funcion para hallar el Maximo Comun Divisor """
def mcd(num1, num2):
    maximo = max(num1, num2)
    minimo = min(num1, num2)
    while minimo!=0:
        mcd = minimo
        minimo = maximo%minimo
        maximo = mcd
    return mcd
 
""" Funcion para hallar el Minimo Comun Multiplo """
def mcm(num1, num2):
    maximo = max(num1, num2)
    minimo = min(num1, num2)
    mcm = (maximo / mcd(maximo, minimo)) * minimo
    return mcm
 
""" Funcion Principal """
# Pedimos los datos al usuario
num1 = int(input("Ingrese el primer numero\n:"))
num2 = int(input("Ingrese el segundo numero\n:"))
# Mostramos el resultado en pantalla
print("El M.C.M. entre "+str(num1)+" y "+str(num2)+" es: "+str(mcm(num1, num2)))
```

5. Explica como combinar 4 gráficos con subplot.

Se utiliza subplots, le tenemos que pasar el número de filas y columnas donde queremos que representar las gráficas, en nuestro ejemplo va a ser una fila y cuatro columnas.
Por otro lado también hay que tener en cuenta el tamaño del conjunto de la figura. En nuestro caso le hemos datos 20 de largo y 4 de ancho.
Una vez creado en el vector axes hay que ir generando las gráficas que queremos dibujar.

2. Cuenta el número de líneas del fichero.

```python
#Realizamos una función que cuenta las lineas de fichero.
def cuentaLineas(file):
  Counter = 0
    
  Content = file.read() 
  CoList = Content.split("\n") 
    
  for i in CoList: 
      if i: 
          Counter += 1
            

  return(Counter) 

file = open("FicheroALeer.txt","r") 
numLineas=cuentaLineas(file)
print("El número de lineas es") 
print(numLineas) 
```

3. Cuenta el número de veces que aparece el artículo «el» en el fichero 

```python
#Realizamos una función que dado un fichero, busca la palabra que deseas.
def buscaPalabra(fichero,palabraObjetivo):
    lineas = []
    contador=0
    #Por cada linea leida vamos incrementando el contador
    for lineas in fichero:
    # separar línea en palabras, por espacio en blanco
        palabras = lineas.split() 
        for palabra in palabras:
        #Pasamos a minusculas la palabra y la comparamos.        
            if palabra.lower() == palabraObjetivo:
                contador=contador+1
    return contador    
#En nuestro caso la palabra es "el" al ser un artículo y no buscamos "él" ya que es un pronombre y el ejercicio sólo pide el artículo.
palabraObjetivo="el"
fichero = open('FicheroALeer.txt', "r")
numeroOcasiones=buscaPalabra(fichero,palabraObjetivo)
print("La palabra buscada " + str(palabraObjetivo) + " se encuentra en " +  str(numeroOcasiones) + " ocasiones.")
fichero.close()

```

4. Elimina las tildes del fichero y convierta todas las letras a minúscula.

```python
import unicodedata

#Creamos una función que elimina las tildes
def eliminaTildesPalabras(cadena):
    s = ''.join((c for c in unicodedata.normalize('NFD',cadena) if unicodedata.category(c) != 'Mn'))
    return s

#Función que elimina tildes y pasa a minusculas
def eliminaTildesPasaMinusculas(fichero):
    lineas = []
    nuevaLineas = []
    contadorLineas=0
    lineaActual=1
    for lineas in fichero:
    # separaramos línea en palabras, por espacio en blanco
        palabras = lineas.split() 
        #aumentamos el contador de lineas
        contadorLineas=contadorLineas+1
        for palabra in palabras:
            
            if lineaActual != contadorLineas:
                #Control para añadir el salto de linea.
                nuevaLineas.append('\n')
                
            #Pasamos a minusculas la palabra
            palabra = palabra.lower()
            #Eliminamos las tildes
            palabra = eliminaTildesPalabras(palabra)
            #Añadimos la palabra tratada una nueva lista
            nuevaLineas.append(palabra + ' ')

            lineaActual=contadorLineas             

    return nuevaLineas


fichero = open('FicheroALeer.txt', "r")
textoTratado=eliminaTildesPasaMinusculas(fichero)
fichero.close()
```

5. Crea un fichero de texto con el nombre «resultado.txt» 

```python
ficheroResultado = open("resultado.txt", "w")
ficheroResultado.write("El numero de lineas es: " + str(numLineas) + "\n")
ficheroResultado.write("La palabra buscada " + str(palabraObjetivo) + " se encuentra en " +  str(numeroOcasiones) + " ocasiones." + "\n")

for palabraTratada in textoTratado:
    
    ficheroResultado.write(str(palabraTratada))

ficheroResultado.close()

```

4. Crea un DataFrame de Pandas donde la primera columna sean todas las provincias que aparecen en el archivo, la segunda columna el número total de habitantes por provincia, la tercera la desviación típica en el número total de habitantes por provincia y la cuarta el número total de municipios por provincia.

```python
#Generamos la primera columna nombre de las provincias
#Sumamos los habitantes por provincia
final=df.groupby('Provincia')['Poblacion'].sum()

#Calculamos la desviación típica del número de habitantes por provincia.
desviacionHabitantesProvincias=df.groupby('Provincia')['Poblacion'].std()

#Añadimos la columna al dataframe final.
final=pd.merge(final, desviacionHabitantesProvincias, on='Provincia')
#Calculamos el numero de municipios por provincia.
totalMunicipios=df.groupby('Provincia')['Municipio'].count()

#Añadimos la columna al dataframe final.
final=pd.merge(final, totalMunicipios, on='Provincia')

#Imprimimos para verificar que esta todo ok.
print(final)
```
