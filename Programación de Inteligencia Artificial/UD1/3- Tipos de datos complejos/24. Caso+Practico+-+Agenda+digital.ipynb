````python
def leer_agenda(nombre_agenda):
  agenda_digital_lectura = open(nombre_agenda, "r")
  agenda_digital = agenda_digital_lectura.readlines()
  agenda_digital_lectura.close()

  return eval(agenda_digital[0])
````

````python
def solicitar_contacto():
  nombre = input("Introduce el nombre del nuevo contacto:")
  direccion = input("Introduce la dirección del nuevo contacto: ")
  email = input("Introduce el correo electrónico del nuevo contacto: ")
  telefono = input("Introduce el número de teléfono del nuevo contacto: ")

  nuevo_contacto = {
    "nombre": nombre,
    "direccion": direccion,
    "email": email,
    "telefono": telefono
  }

  return nuevo_contacto
````

````python
nuevo_contacto = solicitar_contacto()
````

````python
def crear_contacto(agenda_digital, nuevo_contacto):
    agenda_digital[nuevo_contacto['nombre']] = {
        "dirección": nuevo_contacto["direccion"],
        "email": nuevo_contacto["email"],
        "telefono": nuevo_contacto["telefono"]
    }
    print("El contacto se ha creado satisfactoriamente.")

    return agenda_digital
````

````python
crear_contacto(agenda_digital, nuevo_contacto)
````
