### Perceptrn multicapa (MLP)



Un perceptrn multicapa (MLP) es una red neuronal artificial que consiste en mltiples capas de unidades, destinadas a modelar relaciones complejas en los datos. Funciona mediante la composicin de diferentes transformaciones lineales y no lineales de los datos de entrada para realizar tareas de clasificacin o regresin.

**Estructura del MLP:**

*   **Capa de Entrada**: Recibe las seales (caractersticas) del conjunto de datos. No realiza ningn clculo, solo distribuye los datos a la siguiente capa.
    
*   **Capas Ocultas**: Estas capas realizan transformaciones matemticas de las entradas. Cada neurona en estas capas aplica una suma ponderada de sus entradas y luego pasa el resultado a travs de una funcin de activacin no lineal.

**Funciones de Activacin:** Las funciones de activacin no lineales permiten que el MLP aprenda relaciones no lineales complejas. Algunas funciones de activacin comunes son:

*   **Sigmoide**: Produce una salida entre 0 y 1, til para probabilidades.
    
*   **Tanh**: Versin escalada de la funcin sigmoide que produce salidas entre -1 y 1.
    
*   **ReLU (Unidad Lineal Rectificada)**: Da cero para entradas negativas y la misma entrada para entradas positivas. Es la ms utilizada debido a su simplicidad computacional y buen rendimiento.
    

**Entrenamiento del MLP:** El entrenamiento se realiza a travs del algoritmo de retropropagacin (backpropagation) y el descenso de gradiente. Este proceso implica:

1.  Propagacin hacia adelante de las entradas para generar una salida.
2.  Comparacin de la salida con la salida deseada y clculo del error.
3.  Propagacin hacia atrs del error a travs de la red para calcular el gradiente del error respecto a cada peso.
4.  Ajuste de pesos para minimizar el error, generalmente con descenso de gradiente o variantes como Adam o SGD.

**Prevencin de Sobreajuste:** Para evitar el sobreajuste, se pueden utilizar tcnicas como:

*   **Regularizacin (L1, L2)**: Penaliza los pesos grandes.
*   **Dropout**: Desactiva aleatoriamente neuronas durante el entrenamiento para evitar que la red dependa demasiado de cualquier conjunto de neuronas.
*   **Early Stopping**: Detiene el entrenamiento cuando el rendimiento en un conjunto de validacin comienza a empeorar.



## Funcin de activacin
En las redes neuronales artificiales, las funciones de activacin son componentes clave que se utilizan en las neuronas o unidades (nodos) de la red para introducir no linealidad en el modelo. La funcin de activacin determina la salida de una neurona en funcin de la suma ponderada de sus entradas. Bsicamente, toma una entrada y produce una salida basada en cierta regla matemtica. Hay varias funciones de activacin comunes en la inteligencia artificial, y algunas de las ms utilizadas incluyen:

*   Funcin de Activacin Lineal (Linear Activation): Esta funcin simplemente pasa la entrada sin ningn cambio, y es comnmente utilizada en capas de salida de modelos de regresin.
    
*   Funcin Softmax: La funcin softmax se utiliza tpicamente en la capa de salida de redes neuronales para problemas de clasificacin multiclase. Convierte las puntuaciones de entrada en una distribucin de probabilidad, donde la suma de todas las salidas es igual a 1.
