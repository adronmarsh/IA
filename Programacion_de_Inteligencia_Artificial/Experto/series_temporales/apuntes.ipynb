# Series temporales
Las series temporales en el contexto de la Inteligencia Artificial (IA) se refieren a secuencias de datos o puntos que se registran, o se muestren, a lo largo del tiempo. Estos datos son fundamentales en muchos campos, como la economía, la meteorología, la medicina, y la ingeniería, y son cruciales para el análisis predictivo y la modelización en IA. Aquí hay algunos puntos clave sobre las series temporales en IA:

1.  **Definición de Series Temporales**: Una serie temporal es una secuencia de observaciones registradas a intervalos de tiempo regulares o irregulares. Cada punto de datos en una serie temporal tiene una marca de tiempo asociada.
    
2.  **Componentes de una Serie Temporal**: Las series temporales a menudo incluyen componentes como tendencia (un patrón de largo plazo), estacionalidad (patrones que se repiten en intervalos regulares), y ruido (variaciones aleatorias).
    
3.  **Uso en IA**: En IA, las series temporales se utilizan para el análisis predictivo, como pronosticar el mercado de valores, la demanda de productos, o el clima. Los modelos de IA aprenden de los datos históricos y tratan de predecir futuros puntos de datos.
    
4.  **Métodos de Modelización**: Algunos métodos comunes incluyen:
    
    *   **Modelos ARIMA (Autoregressive Integrated Moving Average)**: Utilizados para modelar y predecir series temporales lineales.
    *   **Redes Neuronales Recurrentes (RNN)**: Particularmente efectivas para series temporales debido a su habilidad para mantener información en el tiempo.
    *   **LSTM (Long Short-Term Memory)**: Una variante de RNN que es mejor para capturar dependencias de largo plazo.
    *   **Modelos de Procesamiento de Lenguaje Natural (NLP)**: Como los modelos Transformer, que se han adaptado para el análisis de series temporales.
5.  **Análisis y Preprocesamiento de Datos**: Incluye la limpieza de datos, el manejo de valores faltantes, la normalización, y la detección de estacionalidad y tendencia.
    
6.  **Desafíos en Series Temporales**: Incluyen la dependencia temporal (la correlación entre observaciones), la variabilidad de las series temporales (diferentes patrones o cambios en el tiempo), y la necesidad de grandes cantidades de datos para entrenar modelos precisos.
    
7.  **Aplicaciones en Diversos Campos**: Desde el pronóstico del tiempo y la predicción de precios de acciones hasta el monitoreo de la salud y la predicción de demanda en retail.
    
8.  **Herramientas y Tecnologías**: Existen diversas librerías y herramientas para trabajar con series temporales en IA, como TensorFlow, PyTorch, Keras, y otras específicas como Prophet de Facebook.
    
9.  **Evaluación del Modelo**: Utiliza métricas específicas como el error cuadrático medio (MSE), la raíz del error cuadrático medio (RMSE), y otros indicadores para medir la precisión de las predicciones.


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generando una serie temporal simple
np.random.seed(0)
dates = pd.date_range('2024-01-01', periods=100)
data = np.random.randn(100).cumsum()

# Creando un DataFrame de pandas
df = pd.DataFrame(data, index=dates, columns=['Valor'])

# Graficando la serie temporal
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Valor'], marker='o', linestyle='-', color='blue')
plt.title('Ejemplo de Serie Temporal')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.grid(True)
plt.show()
```

```python
import requests
from datetime import datetime

def obtener_pronostico(api_key, lat, lon):
    url = f"https://api.openweathermap.org/data/2.5/onecall?lat={lat}&lon={lon}&exclude=current,minutely,hourly,alerts&units=metric&appid={api_key}"
    response = requests.get(url)
    
    # Comprobando el código de estado de la respuesta
    if response.status_code == 200:
        return response.json()
    else:
        # Imprimir el código de estado y el mensaje de error
        print(f"Error en la solicitud a la API: Estado {response.status_code}")
        print("Mensaje:", response.text)
        return None

api_key = ''
lat, lon = 40.4168, -3.7038  # Ejemplo: Coordenadas de Madrid, España

pronostico = obtener_pronostico(api_key, lat, lon)
if pronostico:
    for dia in pronostico['daily']:
        fecha = datetime.fromtimestamp(dia['dt']).strftime('%Y-%m-%d')
        temp_min = dia['temp']['min']
        temp_max = dia['temp']['max']
        print(f"Fecha: {fecha}, Temp. Min: {temp_min}°C, Temp. Max: {temp_max}°C")
else:
    print("No fue posible obtener el pronóstico del tiempo.")
```
