```python
# Vamos a cargar nuestro conjunto de datos
iris_dataset = load_iris()
```

```python
# leer el conjunto de datos con Pandas

import pandas as pd
import numpy as np

# nameColumns = iris_dataset['target_names'] + iris_dataset['target']

df = pd.DataFrame(np.c_[iris_dataset["data"], iris_dataset["target"]], columns = iris_dataset['feature_names'] + ['target'])

df
```

```python
# Visualizar los datos

# La 0 es setosa, la 1 versicolor y la 2 

# scatter plt.scatter(x, y, c = b)
# Los pÃ©talos ancho y alto
# de los 3 tipos de flores

import matplotlib.pyplot as plt

fig = plt.figure(figsize = (10, 10))
plt.scatter(df["petal length (cm)"][df["target"] == 0 ],
            df["petal width (cm)"][df["target"] == 0], c = "b", label = "setosa")
plt.scatter(df["petal length (cm)"][df["target"] == 1 ],
            df["petal width (cm)"][df["target"] == 1], c = "r", label = "versicolor")
plt.scatter(df["petal length (cm)"][df["target"] == 2 ],
            df["petal width (cm)"][df["target"] == 2], c = "y", label = "virgÃ­nica")

plt.xlabel("petal_length", fontsize = 14)
plt.ylabel("petal_width", fontsize = 14)
plt.legend(loc = "lower right", fontsize = 14)

plt.show
```

```python
fig = plt.figure(figsize = (15, 15))
ax = plt.axes(projection = "3d")
ax.scatter(df["petal length (cm)"][df["target"] == 0 ],
            df["petal length (cm)"][df["target"] == 0 ],
            df["sepal width (cm)"][df["target"] == 0], c = "b", label = "setosa")
ax.scatter(df["petal length (cm)"][df["target"] == 1 ],
            df["sepal width (cm)"][df["target"] == 1],
            df["petal width (cm)"][df["target"] == 1], c = "r", label = "versicolor")
ax.scatter(df["petal length (cm)"][df["target"] == 2 ],
            df["sepal width (cm)"][df["target"] == 2],
            df["petal width (cm)"][df["target"] == 2], c = "y", label = "virgÃ­nica")

ax.set_xlabel("petal_length")
ax.set_ylabel("petal_width")
ax.set_zlabel("sepal_width")
plt.legend(loc = "best", fontsize = 10)

plt.show
```

```python
# Entrenamiento de nuestro perceptrÃ³n
# Reducir el conjunto de datos
# Los pÃ©talos

df_reduced = df[["petal length (cm)", "petal width (cm)", "target"]]
df_reduced
```

```python
# Separar los datos de la etiqueta

X_df = df_reduced[["petal length (cm)", "petal width (cm)"]]
y_df = df_reduced["target"]
```

```python
from sklearn.linear_model import Perceptron
```

```python
model = Perceptron(max_iter = 1000)
model.fit(X_df, y_df)
```

```python
# RepresentaciÃ³n grÃ¡fica del lÃ­mite de decisiÃ³n
import numpy as np

X = X_df.values

mins = X.min(axis=0) - 0.1
maxs = X.max(axis=0) + 0.1

xx, yy = np.meshgrid(np.linspace(mins[0], maxs[0], 1000),
                    np.linspace(mins[1], maxs[1], 1000))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

fig = plt.figure(figsize=(10, 7))

plt.contour(xx, yy, Z, linewidth = 1, colors = 'r')
```
