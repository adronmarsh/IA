## Navegación vertical (árbol)
la libreria nos permite navegar por la estructura del objeto dom  (árbol) de varias maneras
* utilizando puntos para navegar al elemento interior /hijo
* utilizando la propiedad childern  (.children solo considera los hijos directos de una etiqueta)
* utilizando la propiedad descendants (.descendants considerar los hijos dicertos y los hijos indirectos)
* usando la propiedad parent (.parent considera el elemento padre directo del elemento actual)
* usando la propiedad parents (.parents considera a todos los acendientes)

```python
soup_ordenador.html.body
#soup_ordenador.html.body.div.div
```

```python
tag = soup_ordenador.html.body
for child in tag.children:
    print (child)
```

## Navegación horizontal 

nos permite movernos ente elemento hermanos de una documento html
* .next_sibling  --> siguiente elemento hermano
* .previous_sibling --> anterior elemento hermano
* .next_siblings  -->  lista de los siguientes hermanos
* .previous_siblings --> lista de los anteriores hermanos


```python
sibling_soup = BeautifulSoup( "<a><b>text1</b><c>text2</c><c>text3</c><c>text4</c></b></a>" )
print (sibling_soup.prettify())
```

```python
sibling_soup.b.next_sibling
```

```python
sibling_soup.c.previous_sibling
```

```python
for sibling in sibling_soup.b.next_siblings:
   print(type(sibling)) 
```

## Funciones de búsqueda principales
* find_all . Este método extrae todos los objetos Tag y NavigableString de la
estructura analizada que coinciden con los criterios dados. Función equivalente
(accesible en versiones anteriores de BeautifulSoup): findAll .
find. En este caso, la función solo devuelve el primer objeto que coincide con
los criterios dados.
* find_all_next . Extrae todos los elementos siguientes a un objeto dado, que
cumplen con los criterios especificados. Función equivalente (accesible en
versiones anteriores de BeautifulSoup): findAllNext .
* find_next . Identifica el primer elemento siguiente a un objeto dado, que cumple con los criterios especificados. Función equivalente (accesible en versiones anteriores de BeautifulSoup): findNext .
* find_all_previous . Extrae todos los elementos previos a un objeto dado, que
cumplen con los criterios especificados. Función equivalente (accesible en
versiones anteriores de BeautifulSoup): findAllPrevious .
* find_previous . Identifica el primer elemento previo a un objeto dado, que
cumple con los criterios especificados. Función equivalente (accesible en
versiones anteriores de BeautifulSoup): findPrevious .
* find_next_siblings . Extrae todos los elementos hermanos siguientes de un
objeto dado, que cumplen con los criterios especificados. Función equivalente
(accesible en versiones anteriores de BeautifulSoup): findNextSiblings .
* find_next_sibling . Identifica el primer elemento hermano siguiente de un
objeto dado, que cumple con los criterios especificados. Función equivalente
(accesible en versiones anteriores de BeautifulSoup): findNextSiblin
* find_previous_siblings . Extrae todos los elementos hermanos previos de un
objeto dado, que cumplen con los criterios especificados. Función equivalente
(accesible en versiones anteriores de BeautifulSoup): findPreviousSiblings .
* find_previous_sibling . Identifica el primer elemento hermano previo de un
objeto dado, que cumple con los criterios especificados. Función equivalente
(accesible en versiones anteriores de BeautifulSoup): findPreviousSibling .
* find_parents . Extrae todos los elementos padre de un objeto dado, que
cumplen con los criterios especificados. Función equivalente (accesible en
versiones anteriores de BeautifulSoup): findParents .
* find_parent . Identifica el primer elemento padre de un objeto dado, que
cumple con los criterios especificados. Función equivalente (accesible en
versiones anteriores de BeautifulSoup): findParent .
