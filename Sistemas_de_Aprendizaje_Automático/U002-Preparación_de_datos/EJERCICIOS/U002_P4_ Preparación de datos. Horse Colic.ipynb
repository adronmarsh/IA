## Can you predict the life or death of a horse?

```python
import numpy as np
import pandas as pd


dataset = pd.read_csv('horse.csv', header = 0)
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset['surgery'] = dataset['surgery'].map(lambda s: 1 if s == 'yes' else 0)
dataset.head()
```

```python
dataset['age'] = dataset['age'].map(lambda s: 1 if s == 'adult' else 0)
dataset.head()
```

```python
dataset = dataset.drop('hospital_number', axis= 1)
dataset.head()
```

```python
dataset['rectal_temp'] = dataset.fillna(dataset['rectal_temp'].mode)
dataset.head()
```
